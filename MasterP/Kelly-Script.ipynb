```python
import Kelly2009
import numpy as np
import matplotlib.pyplot as plt
```

```python
MAG_VEGA_B = 0.03
MAG_VEGA_V = 0.03
MAG_VEGA_G = 0.03
MAG_VEGA_J = -0.18
MAG_VEGA_H = -0.03
MAG_VEGA_K = 0.13

MAG_VEGA_g = -0.08
MAG_VEGA_r = 0.16
MAG_VEGA_i = 0.37
MAG_VEGA_z = 0.54

FLUX_LAMBDA_VEGA_B = 632*10**(-11) #erg/cm**2/s/A
FLUX_LAMBDA_VEGA_V = 363.1*10**(-11)
FLUX_LAMBDA_VEGA_G = float('nan')
FLUX_LAMBDA_VEGA_J = 31.47*10**(-11)
FLUX_LAMBDA_VEGA_H = 11.38*10**(-11)
FLUX_LAMBDA_VEGA_K = 3.961*10**(-11)

FLUX_LAMBDA_VEGA_g = 466.9*10**(-11) #erg/cm**2/s/A
FLUX_LAMBDA_VEGA_r = 278.0*10**(-11)
FLUX_LAMBDA_VEGA_i = 185.2*10**(-11)
FLUX_LAMBDA_VEGA_z = 131.5*10**(-11)
```

```python
data = np.loadtxt('NOVEMBER/NOV-NGC3783-K')
error = np.loadtxt('NOVEMBER/NGC3783_NOISE_K.txt')

#print data[:,0]
#print data[:,1]
#print error[:,0]
#print error[:,4]

time2_K = data[:,0]
flux2_K = data[:,1]
sigma2_K = error[:,1]

time1_K = []
flux1_K = []
sigma1_K = []

for i in range(len(time2_K)):
    if np.isnan(sigma2_K[i]):
        pass
    elif np.isnan(flux2_K[i]):
        pass
    else:
        time1_K.append(time2_K[i])
        flux1_K.append(flux2_K[i])
        sigma1_K.append(sigma2_K[i])

time1_K = np.array(time1_K)
flux1_K = np.array(flux1_K)
sigma1_K = np.array(sigma1_K)

time_K = np.insert(time1_K,0,0)
sigma_K = np.insert(sigma1_K,0,np.mean(sigma1_K))
flux_K = np.insert(flux1_K,0,np.mean(flux1_K))

sigma1_K = -2.5*np.log10((flux1_K + sigma1_K)/FLUX_LAMBDA_VEGA_K)
sigma_K = -2.5*np.log10((flux_K + sigma_K)/FLUX_LAMBDA_VEGA_K)
flux1_K = -2.5*np.log10(flux1_K/FLUX_LAMBDA_VEGA_K)
flux_K = -2.5*np.log10(flux_K/FLUX_LAMBDA_VEGA_K)

sigma1_K = abs(sigma1_K - flux1_K)
sigma_K = abs(sigma_K - flux_K)

global_variable_share = np.array((time_K,flux_K,sigma_K))
global_variable_share_2 = np.array((time1_K,flux1_K,sigma1_K))

np.savetxt('timefluxsigma.txt',global_variable_share)
np.savetxt('time1flux1sigma1.txt',global_variable_share_2)

reload(Kelly2009)
from Kelly2009 import *
param = [0.01,800.,0.01]
model_K = flux_model(param)
#print model_K
np.save('NOVEMBER/Kelly2009model-NGC3783_K',model_K)
```
