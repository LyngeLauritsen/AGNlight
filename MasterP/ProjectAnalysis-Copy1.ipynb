```python
def dist(mid,end):
    return np.sqrt((mid[0] - end[0])**2 + (mid[1] - end[1])**2)
```

```python
def Basic(quasar,header):
    '''Provides a basic list of neccesities from the fits header file for use in the later flux determination, \
    it does not serve an individual purpose beyound that'''
    AGNdata = quasar #np.rot90(np.rot90(np.fliplr(quasar)))
    #header = data[0].header
    #print header
    RAstd = header['CRVAL1']
    DECstd = header['CRVAL2']
    #print RAstd, DECstd
    pixspa = header['PIXSCALE']/(3600.)
    #print pixspa
    RAstdpix = header['CRPIX1']
    DECstdpix = header['CRPIX2']
    exptime = header['EXPTIME']
    return header,AGNdata,RAstd,DECstd,pixspa,RAstdpix,DECstdpix,exptime
```

```python
def curve(quasar,header,w,center,apparature):
    '''Determines the FLUX of the stellar object given the the numpy array of the image, \
    the header of the fits file, the astropy coordinate representation, \
    the object position and the apparature of interest'''
    header,AGNdata,RAstd,DECstd,pixspa,RAstdpix,DECstdpix,exptime = Basic(quasar,header)
    #AGNdata = np.swapaxes(AGNdata,0,1)
    y, x = np.ogrid[0:header['NAXIS1'],0:header['NAXIS1']]
    #w = np.swapaxes(w,0,1)
    #F_AGN = np.zeros((512,512,3)) #((int(2*apparature/pixspa)+1,int(2*apparature/pixspa)+1,3))
    #print center
    x1 = w.all_world2pix(center[0],center[1],0)
    x1,y1 = (x1[0]),(x1[1])
    mask = ((y-y1)**2 + (x-x1)**2) > (apparature/float(pixspa))**2
    AGNdata[mask] = float(0)
    mask = AGNdata == 0
    rows = np.flatnonzero((~mask).sum(axis=1))
    cols = np.flatnonzero((~mask).sum(axis=0))
    AGNdata = AGNdata[rows.min():rows.max()+1, cols.min():cols.max()+1]
    #print apparature/float(pixspa)
    return AGNdata/exptime, x1, y1
```

```python
def SXtable(quas,aperture):
    flux1_radq = []
    flux1_rads = []
    flux1_comp = []
    time = []
    err = []
    err_bar = []
    posx = []
    posx_err = []
    posy = []
    posy_err = []
    for i in range(len(quas)):
        print i
        s = i
        #print header
        header = pyfits.getheader(quas[s])
        phot_app = aperture*3600/header['PIXSCALE']
        #print phot_app
        np.set_printoptions(threshold=np.inf)
        sew = sewpy.SEW(params=['XMODEL_WORLD', 'YMODEL_WORLD', 'X2MODEL_WORLD', 'Y2MODEL_WORLD', 
                                'XMODEL_IMAGE', 'YMODEL_IMAGE', 'FLUX_RADIUS(1)', 'FLUX_APER'],
                        config={"DETECT_THRESH":5, 'DETECT_MINAREA':5, 'DETECT_MAXAREA':100, 
                                "PHOT_FLUXFRAC":"0.5", 'PHOT_APERTURES':2*phot_app, 
                                'PIXEL_SCALE':header['PIXSCALE']}
                        , sexpath='/usr/bin/sextractor')
        out = sew(quas[s])
        #print out['FHWM_WORLD']
        return out
```

```python
def flux(quas,aperture,main_RA_WORLD,main_DEC_WORLD,sec_RA_WORLD,sec_DEC_WORLD):
    #print out['table'] #['FWHM_WORLD'][:]
    #print ' ' 
    #print RA_WORLD,DEC_WORLD, WCS(quas[s]).all_world2pix(RA_WORLD,DEC_WORLD,0)
    RAmain = WCS(quas[s]).all_world2pix(main_RA_WORLD,main_DEC_WORLD,1)[0]
    DECmain = WCS(quas[s]).all_world2pix(main_RA_WORLD,main_DEC_WORLD,1)[1]
    RAsec = WCS(quas[s]).all_world2pix(sec_RA_WORLD,sec_DEC_WORLD,1)[0]
    DECsec = WCS(quas[s]).all_world2pix(sec_RA_WORLD,sec_DEC_WORLD,1)[1]
    print RAmain, DECmain
    #print 'Should be', sta2_RA_WORLD, sta2_DEC_WORLD, RAsta2, DECsta2
    #print ' ' 
    #print ' '
    #print ' ' 
    #print '174.77418897 -37.753703989', WCS(quas[i]).all_world2pix(174.77418897,-37.753703989,0)
    #print len(out)
    flux_radq = []
    flux_rads = []
    out = SXtable(quas,aperture)
    #print RAsta, DECsta, RAqua, DECqua
    for j in range(len(out['table']['XMODEL_IMAGE'])):
        #print out['table']['X_WORLD'][j]
        #print WCS(quas[i]).all_world2pix(RA_WORLD,DEC_WORLD,0)[0]
        if RAmain - 1. < out['table']['XMODEL_IMAGE'][j] <  RAmain + 1. and \
        DECmain - 1. < out['table']['YMODEL_IMAGE'][j] < DECmain + 1.:
            for k in range(len(out['table']['XMODEL_IMAGE'])):
                if RAsec - 1. < out['table']['XMODEL_IMAGE'][k] <  RAsec + 1. and \
                DECsec - 1. < out['table']['YMODEL_IMAGE'][k] < DECsec + 1.:
                    #print out['table']['X_IMAGE'][j], j
                    #print out['table']['X_IMAGE'][k], k
                    print out['table']['FLUX_RADIUS'][j]
                    flux1_radq.append(out['table']['FLUX_APER'][j])
                    flux1_rads.append(out['table']['FLUX_APER'][k])
                    flux1_comp.append(out['table']['FLUX_APER'][j]/out['table']['FLUX_APER'][k])
                    time.append(header['MJD-OBS'])
                    err.append(3600*np.sqrt((out['table']['XMODEL_WORLD'][j] - main_RA_WORLD)**2 \
                                            + (out['table']['YMODEL_WORLD'][j] - main_DEC_WORLD)**2))
                    err_bar.append(np.sqrt(out['table']['X2MODEL_WORLD'][j]**2 \
                                           + out['table']['Y2MODEL_WORLD'][j]**2))
                    posx.append(out['table']['XMODEL_WORLD'][j])
                    posx_err.append(out['table']['X2MODEL_WORLD'][j])
                    posy.append(out['table']['YMODEL_WORLD'][j])
                    posy_err.append(out['table']['Y2MODEL_WORLD'][j])
    #flux1_radq.append(flux_radq)
    #flux1_rads.append(flux_rads)
return flux1_radq, flux1_rads, flux1_comp, time, err, err_bar, posx, posx_err, posy, posy_err
```
