```python
import seaborn
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import os
import pyfits
```

```python
SNR_K = np.loadtxt('../NOVEMBER/NGC7213_SNR_Ks7.txt')
SNR_H = np.loadtxt('../NOVEMBER/NGC7213_SNR_Hs7.txt')
SNR_J = np.loadtxt('../NOVEMBER/NGC7213_SNR_Js7.txt')
SNR_g = np.loadtxt('../NOVEMBER/NGC7213_SNR_gs7.txt')
SNR_r = np.loadtxt('../NOVEMBER/NGC7213_SNR_rs7.txt')
SNR_i = np.loadtxt('../NOVEMBER/NGC7213_SNR_is7.txt')
SNR_z = np.loadtxt('../NOVEMBER/NGC7213_SNR_zs7.txt')

print 'SNR g =', np.nanmean(SNR_g[:,1]), np.nanstd(SNR_g[:,1])
print 'SNR r =', np.nanmean(SNR_r[:,1]), np.nanstd(SNR_r[:,1])
print 'SNR i =', np.nanmean(SNR_i[:,1]), np.nanstd(SNR_i[:,1])
print 'SNR z =', np.nanmean(SNR_z[:,1]), np.nanstd(SNR_z[:,1])
print 'SNR J =', np.nanmean(SNR_J[:,1]), np.nanstd(SNR_J[:,1])
print 'SNR H =', np.nanmean(SNR_H[:,1]), np.nanstd(SNR_H[:,1])
print 'SNR K =', np.nanmean(SNR_K[:,1]), np.nanstd(SNR_K[:,1])
```

```python
plt.figure(figsize=(15,20))
matplotlib.rc('xtick', labelsize=16) 
matplotlib.rc('ytick', labelsize=16) 

plt.subplot(7,1,1)
plt.scatter(SNR_g[:,0],SNR_g[:,1],color='yellow',s=50,label='$g$ Band')
plt.scatter(SNR_g[:,0],SNR_g[:,1],color='orange',s=50,label='$r$ Band')
plt.scatter(SNR_g[:,0],SNR_g[:,1],color='purple',s=50,label='$i$ Band')
plt.scatter(SNR_g[:,0],SNR_g[:,1],color='brown',s=50,label='$z$ Band')
plt.scatter(SNR_g[:,0],SNR_g[:,1],color='r',s=50,label='$J$ Band')
plt.scatter(SNR_g[:,0],SNR_g[:,1],color='g',s=50,label='$H$ Band')
plt.scatter(SNR_g[:,0],SNR_g[:,1],color='b',s=50,label='$K$ Band')
plt.scatter(SNR_g[:,0],SNR_g[:,1],color='yellow')
plt.ylabel('SNR',fontsize=20)
#plt.scatter(float('nan'),float('nan'),color='orange',label='r Band')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.40),
          ncol=4, fancybox=True, shadow=True,fontsize=18)
plt.grid('off')
plt.xticks([])

plt.subplot(7,1,2)
plt.scatter(SNR_r[:,0],SNR_r[:,1],color='orange',label='r Band')
plt.ylabel('SNR',fontsize=20)
plt.grid('off')
plt.xticks([])

plt.subplot(7,1,3)
plt.scatter(SNR_i[:,0],SNR_i[:,1],color='purple',label='i Band')
plt.ylabel('SNR',fontsize=20)
plt.grid('off')
plt.xticks([])

plt.subplot(7,1,4)
plt.scatter(SNR_z[:,0],SNR_z[:,1],color='brown',label='z Band')
plt.ylabel('SNR',fontsize=20)
plt.grid('off')
plt.xticks([])

plt.subplot(7,1,5)
plt.scatter(SNR_J[:,0],SNR_J[:,1],color='r',label='J Band')
plt.ylabel('SNR',fontsize=20)
plt.grid('off')
plt.xticks([])

plt.subplot(7,1,6)
plt.scatter(SNR_H[:,0],SNR_H[:,1],color='g',label='H Band')
plt.ylabel('SNR',fontsize=20)
plt.grid('off')
plt.xticks([])

plt.subplot(7,1,7)
plt.scatter(SNR_K[:,0],SNR_K[:,1],color='b',label='K Band')
plt.subplots_adjust(wspace=0,hspace=0.05)
plt.ylabel('SNR',fontsize=20)
plt.xlabel('MJD-OBS',fontsize=20)
plt.grid('off')
plt.title('NGC7213 $signal$-$to$-$noise$ $ratio$',fontsize=24,y=7.70)

plt.savefig('/home/lynge/MasterP/paper_here/Figure/NGC7213_SNR.png',bbox_inches='tight')
plt.show()
```
